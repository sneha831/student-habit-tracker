```python
import pandas as pd #We use Pandas to create and manage tables
from datetime import date#To automatically record today's date in the log
```

```python
#Define a Sample Habit Entry
habit_entry = {
    "Name": "Sneha",
    "Date": date.today(),
    "Study Hours": 3,
    "Sleep Hours": 7,
    "Screen Time (hrs)": 5,
    "Mood": "Happy",
    "Stress Level (1-5)": 2
}

```

```python
# Convert to DataFrame
df = pd.DataFrame([habit_entry])#Turns your dictionary into a table (DataFrame)

# Save to CSV
df.to_csv("habit_log.csv", index=False)#Prevents it from saving row numbers (we donâ€™t need them)

# Show the DataFrame
df
```

```python
# Import necessary again if you're in a new cell
import pandas as pd
from datetime import date, timedelta

# Load existing CSV
df_existing = pd.read_csv("habit_log.csv")

# Create a new fake entry for the next day
new_entry = {
    "Name": "Sneha",
    "Date": date.today() + timedelta(days=1),
    "Study Hours": 4,
    "Sleep Hours": 6,
    "Screen Time (hrs)": 4,
    "Mood": "Tired",
    "Stress Level (1-5)": 3
}

# Convert to DataFrame and append
df_new = pd.DataFrame([new_entry])
df_updated = pd.concat([df_existing, df_new], ignore_index=True)

# Save it back to the same file
df_updated.to_csv("habit_log.csv", index=False)

# Show updated file
df_updated
```

```python
import random#This imports Pythonâ€™s random module so we can randomly generate numbers (like hours or stress level) and pick random moods.

# Load existing or create empty list
entries = []

# Generate 5 days of logs
for i in range(5):
    entry = {
        "Name": "Sneha",
        "Date": date.today() + timedelta(days=i),
        "Study Hours": random.randint(2, 6),
        "Sleep Hours": random.randint(5, 9),
        "Screen Time (hrs)": random.randint(3, 7),
        "Mood": random.choice(["Happy", "Tired", "Stressed", "Motivated"]),
        "Stress Level (1-5)": random.randint(1, 5)
    }
    entries.append(entry)

# Convert to DataFrame
df_fake = pd.DataFrame(entries)

# Save (or append) to file
df_fake.to_csv("habit_log.csv", index=False)

# Show result
df_fake
```

```python
import pandas as pd# This simply reloads the file and shows your full table.

# Load the habit log CSV
df = pd.read_csv("habit_log.csv")

# Show the full table
df
```

```python
#Convert Date Column to Date Format
df["Date"] = pd.to_datetime(df["Date"])#df["Date"] = pd.to_datetime(df["Date"])

```

```python
# Line Chart â€“ Study Hours, Sleep Hours, Stress Level
import matplotlib.pyplot as plt

# Set the figure size
plt.figure(figsize=(12,6))

# Plot Study, Sleep, Stress over time
plt.plot(df["Date"], df["Study Hours"], label="Study Hours", marker='o')
plt.plot(df["Date"], df["Sleep Hours"], label="Sleep Hours", marker='o')
plt.plot(df["Date"], df["Stress Level (1-5)"], label="Stress Level", marker='o')

# Titles and labels
plt.title("Daily Study, Sleep, and Stress Levels")
plt.xlabel("Date")
plt.ylabel("Hours / Level")
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)

# Show the plot
plt.tight_layout()
plt.show()
```

```python
#Mood Analysis (Bar Chart)
import matplotlib.pyplot as plt
import seaborn as sns

# Set style
sns.set_style("whitegrid")

# Count how many times each mood appeared
mood_counts = df_fake["Mood"].value_counts()

# Plot bar chart
plt.figure(figsize=(8,5))
sns.barplot(x=mood_counts.index, y=mood_counts.values, palette="pastel")

plt.title("Mood Frequency Over 5 Days")
plt.xlabel("Mood")
plt.ylabel("Count")
plt.show()
```

```python
#Code for Correlation Heatmap:
import seaborn as sns
import matplotlib.pyplot as plt

# Select only the numeric columns
numeric_data = df_fake[["Study Hours", "Sleep Hours", "Screen Time (hrs)", "Stress Level (1-5)"]]

# Compute correlation matrix
corr_matrix = numeric_data.corr()

# Plot heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", center=0)
plt.title("Correlation Heatmap: Habits & Stress")
plt.show()
```
